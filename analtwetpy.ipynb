```python
def limit_handled(cursor):
    while True:
        try:
            yield cursor.next()
        except tweepy.RateLimitError:
            time.sleep( 15*60 )
            continue
        except StopIteration:
            break

                  
diccionario = {}
for follower in limit_handled( tweepy.Cursor(api.followers).items() ):
    
    if follower.friends_count < 300:
        
        diccionario[follower.screen_name] = list()
        
        user = api.get_user( follower.screen_name )
    
        for friend in user.friends():
            
            if friend.friends_count < 300 and friend.screen_name != '_Huguez':
                
                diccionario[follower.screen_name].append( friend.screen_name )
```
